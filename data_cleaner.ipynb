```python
import pandas as pd
import json
```

```python
def load_jsonl_safely(filepath):
    data = []

    with open(filepath, "r", encoding="utf-8") as f:
        for line in f:
            obj = json.loads(line)
            data.append(obj)

    return pd.DataFrame(data)

df = load_jsonl_safely("Content/all_canada_rag_content.jsonl")
```
